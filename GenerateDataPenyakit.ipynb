```python
import os
import xml.etree.ElementTree as ET

def extract_abstract_from_file(file_path):
    try:
        tree = ET.parse(file_path)
        root = tree.getroot()

        # Mencari elemen abstract dalam XML
        abstract_element = root.find(".//AbstractText")

        if abstract_element is not None:
            abstract_text = abstract_element.text
            return abstract_text
        else:
            print(f"Tidak dapat menemukan elemen AbstractText dalam file {file_path}")
            return None
    except ET.ParseError as e:
        print(f"XML Parse Error: {e}")
        return None

def main():
    xml_directory = 'path/to/save/xml/'  # Sesuaikan dengan direktori tempat Anda menyimpan file-file XML

    # Mendapatkan daftar file XML dalam direktori
    xml_files = [f for f in os.listdir(xml_directory) if f.endswith(".xml")]

    for xml_file in xml_files:
        file_path = os.path.join(xml_directory, xml_file)
        abstract_text = extract_abstract_from_file(file_path)

        if abstract_text:
            print(f"Abstract from {xml_file}:\n{abstract_text}\n")

if __name__ == "__main__":
    main()
```

```python
import os
import xml.etree.ElementTree as ET
import requests

def get_article_details(api_key, pmid, save_path):
    base_url = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi"

    params = {
        'db': 'pubmed',
        'id': pmid,
        'retmode': 'xml',
        'api_key': api_key
    }

    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()  # Check for errors in the response

        if response.text.strip():  # Check if response content is non-empty
            # Save XML to file
            save_xml_to_file(response.text, os.path.join(save_path, f'{pmid}.xml'))

            # Extract comorbidity names
            comorbidities = extract_comorbidities(response.text)
            print(f"Comorbidities for PMID {pmid}: {', '.join(comorbidities)}")
        else:
            print("Empty response content")
    except requests.exceptions.HTTPError as errh:
        print("HTTP Error:", errh)
    except requests.exceptions.ConnectionError as errc:
        print("Error Connecting:", errc)
    except requests.exceptions.Timeout as errt:
        print("Timeout Error:", errt)
    except requests.exceptions.RequestException as err:
        print("Error:", err)

    return None

def save_xml_to_file(xml_content, file_path):
    try:
        os.makedirs(os.path.dirname(file_path), exist_ok=True)  # Create directory if not exists
        with open(file_path, 'w', encoding='utf-8') as file:
            file.write(xml_content)
        print(f"XML saved to: {file_path}")
    except Exception as e:
        print(f"Error saving XML: {e}")

def extract_comorbidities(xml_content):
    comorbidities = []
    try:
        root = ET.fromstring(xml_content)
        for element in root.findall('.//MeshHeadingList/MeshHeading/DescriptorName'):
            comorbidity = element.text
            comorbidities.append(comorbidity)
    except ET.ParseError as e:
        print(f"Error parsing XML: {e}")
    return comorbidities

def main():
    api_key = '9edd38677707c08b2a96ed1a6e8c61c47408'  # Replace with your PubMed API key
    query = 'lung cancer AND comorbidities'  # Adjust your search query
    save_path = 'path/to/save/xml/'  # Adjust the path to save XML files

    # Get list of IDs
    id_list =['38112578', '38103121', '38103006', '38102837', '38100106', '38097480', '38092558', '38090511', '38090326', '38090298', '38085543', '38075997', '38075485', '38071660', '38067372', '38067306', '38066848', '38066463', '38063366', '38063017', '38056855', '38056852', '38037323', '38029831', '38029007', '38027239', '38025819', '38023992', '38023743', '38021960', '38016200', '38016003', '38012569', '38011678', '38007507', '38003929', '37999149', '37999105', '37997884', '37996867', '37995046', '37990203', '37985822', '37983029', '37981476', '37979368', '37977487', '37974211', '37973983', '37970369']

    for pmid in id_list:
        get_article_details(api_key, pmid, save_path)

if __name__ == "__main__":
    main()
```
